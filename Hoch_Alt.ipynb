```python
#the number of 0 dimensional face of Hoch(m,n)
def f_0(m,n):
    return 1

#the number of 1 dimensional face of Hoch(m,n)
def f_1(m,n) :
    lst = []
    for i in range(1,m+1) :
        lst.append(i)
    for i in range(n) :
        lst.append(0)
    count = 0
    num = int((m+n)/2)
    for i in range(1,num) :
        comb = list(combinations(lst,2*i))
        count += len(set(comb))
        #print(comb)
    return count

#the number of 2 dimensional face of Hoch(m,n)
def f_2(m,n) :
    lst = []
    for i in range(1,m+1) :
        lst.append(i)
    for i in range(n) :
        lst.append(0)
    count = 0
    num = int((m+n)/2)
    for i in range(1,num-1) :
        comb = list(combinations(lst,2*i))
        for x in set(comb) :
            lst_2 = list(lst)
            for y in x :
                lst_2.remove(y)
            for j in range(1,num-(i)) :
                comb_2 = list(combinations(lst_2,2*j))
                count += len(set(comb_2))
    return count

#the number of 3 dimensional face of Hoch(m,n)
def f_3(m,n) :
    lst = []
    for i in range(1,m+1) :
        lst.append(i)
    for i in range(n) :
        lst.append(0)
    count = 0
    num = int((m+n)/2)
    for i in range(1,num-1) :
        comb = list(combinations(lst,2*i))
        for x in set(comb) :
            lst_2 = list(lst)
            for y in x :
                lst_2.remove(y)
            for j in range(1,num-(i)) :
                comb_2 = list(combinations(lst_2,2*j))
                for z in set(comb_2) :
                    lst_3 = list(lst_2)
                    for w in z :
                        lst_3.remove(w)
                    for k in range(1,num-(i+j)) :
                        comb_3 = list(combinations(lst_3,2*k))
                        count += len(set(comb_3))
    return count

#the number of 4 dimensional face of Hoch(m,n)
def f_4(m,n) :
    if (m+n)%2 == 1:
        print('m+n이 짝수가 아님')
    lst = []
    for i in range(1,m+1) :
        lst.append(i)
    for i in range(n) :
        lst.append(0)
    count = 0
    num = int((m+n)/2)
    for i in range(1,num-1) :
        comb = list(combinations(lst,2*i))
        for x in set(comb) :
            lst_2 = list(lst)
            for y in x :
                lst_2.remove(y)
            for j in range(1,num-(i)) :
                comb_2 = list(combinations(lst_2,2*j))
                for z in set(comb_2) :
                    lst_3 = list(lst_2)
                    for w in z :
                        lst_3.remove(w)
                    for k in range(1,num-(i+j)) :
                        comb_3 = list(combinations(lst_3,2*k))
                        for a in set(comb_3) :
                            lst_4 = list(lst_3)
                            for b in a :
                                lst_4.remove(b)
                            for l in range(1,num-(i+j+k)) :
                                comb_4 = list(combinations(lst_4,2*l))
                                count += len(set(comb_4))
    return count

#the number of 5 dimensional face of Hoch(m,n)
def f_5(m,n) :
    if (m+n)%2 == 1:
        print('m+n이 짝수가 아님')
    lst = []
    for i in range(1,m+1) :
        lst.append(i)
    for i in range(n) :
        lst.append(0)
    count = 0
    num = int((m+n)/2)
    for i in range(1,num-1) :
        comb = list(combinations(lst,2*i))
        for x in set(comb) :
            lst_2 = list(lst)
            for y in x :
                lst_2.remove(y)
            for j in range(1,num-(i)) :
                comb_2 = list(combinations(lst_2,2*j))
                for z in set(comb_2) :
                    lst_3 = list(lst_2)
                    for w in z :
                        lst_3.remove(w)
                    for k in range(1,num-(i+j)) :
                        comb_3 = list(combinations(lst_3,2*k))
                        for a in set(comb_3) :
                            lst_4 = list(lst_3)
                            for b in a :
                                lst_4.remove(b)
                            for l in range(1,num-(i+j+k)) :
                                comb_4 = list(combinations(lst_4,2*l))
                                for c in set(comb_4) :
                                    lst_5 = list(lst_4)
                                    for d in c :
                                        lst_5.remove(d)
                                    for u in range(1,num-(i+j+k+l)) :
                                        comb_5 = list(combinations(lst_5,2*u))
                                        count += len(set(comb_5))
    return count

#the number of 6 dimensional face of Hoch(m,n)
def f_6(m,n) :
    if (m+n)%2 == 1:
        print('m+n이 짝수가 아님')
    lst = []
    for i in range(1,m+1) :
        lst.append(i)
    for i in range(n) :
        lst.append(0)
    count = 0
    num = int((m+n)/2)
    for i in range(1,num-1) :
        comb = list(combinations(lst,2*i))
        for x in set(comb) :
            lst_2 = list(lst)
            for y in x :
                lst_2.remove(y)
            for j in range(1,num-(i)) :
                comb_2 = list(combinations(lst_2,2*j))
                for z in set(comb_2) :
                    lst_3 = list(lst_2)
                    for w in z :
                        lst_3.remove(w)
                    for k in range(1,num-(i+j)) :
                        comb_3 = list(combinations(lst_3,2*k))
                        for a in set(comb_3) :
                            lst_4 = list(lst_3)
                            for b in a :
                                lst_4.remove(b)
                            for l in range(1,num-(i+j+k)) :
                                comb_4 = list(combinations(lst_4,2*l))
                                for c in set(comb_4) :
                                    lst_5 = list(lst_4)
                                    for d in c :
                                        lst_5.remove(d)
                                    for u in range(1,num-(i+j+k+l)) :
                                        comb_5 = list(combinations(lst_5,2*u))
                                        for e in set(comb_5) :
                                            lst_6 = list(lst_5)
                                            for r in e :
                                                lst_6.remove(r)
                                            for v in range(1,num-(i+j+k+l+u)) :
                                                comb_6 = list(combinations(lst_6,2*v))
                                                count += len(set(comb_6))
    return count

#the number of 7 dimensional face of Hoch(m,n)
def f_7(m,n) :
    if (m+n)%2 == 1:
        print('m+n이 짝수가 아님')
    lst = []
    for i in range(1,m+1) :
        lst.append(i)
    for i in range(n) :
        lst.append(0)
    count = 0
    num = int((m+n)/2)
    for i in range(1,num-1) :
        comb = list(combinations(lst,2*i))
        for x in set(comb) :
            lst_2 = list(lst)
            for y in x :
                lst_2.remove(y)
            for j in range(1,num-(i)) :
                comb_2 = list(combinations(lst_2,2*j))
                for z in set(comb_2) :
                    lst_3 = list(lst_2)
                    for w in z :
                        lst_3.remove(w)
                    for k in range(1,num-(i+j)) :
                        comb_3 = list(combinations(lst_3,2*k))
                        for a in set(comb_3) :
                            lst_4 = list(lst_3)
                            for b in a :
                                lst_4.remove(b)
                            for l in range(1,num-(i+j+k)) :
                                comb_4 = list(combinations(lst_4,2*l))
                                for c in set(comb_4) :
                                    lst_5 = list(lst_4)
                                    for d in c :
                                        lst_5.remove(d)
                                    for u in range(1,num-(i+j+k+l)) :
                                        comb_5 = list(combinations(lst_5,2*u))
                                        for e in set(comb_5) :
                                            lst_6 = list(lst_5)
                                            for r in e :
                                                lst_6.remove(r)
                                            for v in range(1,num-(i+j+k+l+u)) :
                                                comb_6 = list(combinations(lst_6,2*v))
                                                for gg in set(comb_6) :
                                                    lst_7 = list(lst_6)
                                                    for h in gg :
                                                        lst_7.remove(h)
                                                    for s in range(1,num-(i+j+k+l+u+v)) :
                                                        comb_7 = list(combinations(lst_7,2*s))
                                                        count += len(set(comb_7))
    return count

#the number of 8 dimensional face of Hoch(m,n)
def f_8(m,n) :
    if (m+n)%2 == 1:
        print('m+n이 짝수가 아님')
    lst = []
    for i in range(1,m+1) :
        lst.append(i)
    for i in range(n) :
        lst.append(0)
    count = 0
    num = int((m+n)/2)
    for i in range(1,num-1) :
        comb = list(combinations(lst,2*i))
        for x in set(comb) :
            lst_2 = list(lst)
            for y in x :
                lst_2.remove(y)
            for j in range(1,num-(i)) :
                comb_2 = list(combinations(lst_2,2*j))
                for z in set(comb_2) :
                    lst_3 = list(lst_2)
                    for w in z :
                        lst_3.remove(w)
                    for k in range(1,num-(i+j)) :
                        comb_3 = list(combinations(lst_3,2*k))
                        for a in set(comb_3) :
                            lst_4 = list(lst_3)
                            for b in a :
                                lst_4.remove(b)
                            for l in range(1,num-(i+j+k)) :
                                comb_4 = list(combinations(lst_4,2*l))
                                for c in set(comb_4) :
                                    lst_5 = list(lst_4)
                                    for d in c :
                                        lst_5.remove(d)
                                    for u in range(1,num-(i+j+k+l)) :
                                        comb_5 = list(combinations(lst_5,2*u))
                                        for e in set(comb_5) :
                                            lst_6 = list(lst_5)
                                            for r in e :
                                                lst_6.remove(r)
                                            for v in range(1,num-(i+j+k+l+u)) :
                                                comb_6 = list(combinations(lst_6,2*v))
                                                for gg in set(comb_6) :
                                                    lst_7 = list(lst_6)
                                                    for h in gg :
                                                        lst_7.remove(h)
                                                    for s in range(1,num-(i+j+k+l+u+v)) :
                                                        comb_7 = list(combinations(lst_7,2*s))
                                                        for p in set(comb_7) :
                                                            lst_8 = list(lst_7)
                                                            for q in p :
                                                                lst_8.remove(q)
                                                            for T in range(1,num-(i+j+k+l+u+v+s)) :
                                                                comb_8 = list(combinations(lst_8,2*T))
                                                                count += len(set(comb_8))
    return count
```

```python
#the Betti numbers of Hoch(m,n) by using the Euler charateristic of induced subcomplex

def g(m,n) :
    P = [1]
    for i in range((m+n)//2) :
        P.append(0)
    for i in range(m+1) :
        a = len(list(combinations(range(m),i)))
        for j in range(n+1) :
            if (i+j)>2 and (i+j)%2 == 0 and j <i+3:
                #print(i,j,a,(-1)**((i+j)//2+1))
                count = 0
                for k in range(1,(i+j)//2) :
                    count += eval('f_{}({},{})'.format(k,i,j))*(-1)**(k)
                P[(i+j)//2] += a*(abs(count)+(-1)**((i+j)//2+1))
            elif i+j == 2 :
                P[(i+j)//2] += a*(eval('f_{}({},{})'.format((i+j)//2-1,i,j)))            
    return P
```
